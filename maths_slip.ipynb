```python
from sympy import Point, Matrix

P = Point(-4, 5)

# a) Scaling in x by factor 2
P1 = P.transform(Matrix([[2, 0, 0], [0, 1, 0], [0, 0, 1]]))

# b) Reflection through line y = x
P2 = P.transform(Matrix([[0, 1, 0], [1, 0, 0], [0, 0, 1]]))

# c) Translation in y by +7
P3 = P.transform(Matrix([[1, 0, 0], [0, 1, 7], [0, 0, 1]]))

print("Scaling (x2):", P1)
print("Reflection (y=x):", P2)
print("Translation (+7 in y):", P3)
```

```python
# Q2
import matplotlib.pyplot as plt
import numpy as np

# Original points
points = np.array([[1,1], [-5,1], [-5,-4], [1,-4], [1,1]])

# Rotation 90° (about origin)
rotation = np.array([[0, -1], [1, 0]])
rotated = np.dot(points, rotation)

# Shearing (x-direction, factor = 3)
shear = np.array([[1, 3], [0, 1]])
sheared = np.dot(points, shear)

# Plot
plt.plot(points[:,0], points[:,1], label="Original")
plt.plot(rotated[:,0], rotated[:,1], label="Rotated 90°")
plt.plot(sheared[:,0], sheared[:,1], label="Sheared")

plt.legend()
plt.title("Polygon Transformations")
grid(True)
plt.show()
```
